## 1) Projeyi getir

**Seçenek A — GitHub:** `REPO_URL` değerini kendi repona göre değiştir.

```python
import os
import subprocess

CONTENT = "/content"
REPO_URL = "https://github.com/mervegulec2/privproject.git"
BRANCH = "main"
PROJECT_ROOT = os.path.join(CONTENT, "privproject")

os.chdir(CONTENT)
if not os.path.isdir(PROJECT_ROOT):
    subprocess.check_call(["git", "clone", REPO_URL, "privproject"])
    os.chdir(PROJECT_ROOT)
    subprocess.check_call(["git", "checkout", BRANCH])
else:
    os.chdir(PROJECT_ROOT)
    subprocess.check_call(["git", "pull"])

os.environ["PROJECT_ROOT"] = os.getcwd()
print("PROJECT_ROOT:", os.environ["PROJECT_ROOT"])
```

## 3) Ortam (.env) ve Dummy Defenses Özelleştirmeleri

Burada Ray Tracing kapalı ve standart Dummy Prototype savunma konfigürasyonu enjekte edilmiştir.
Phase 2/3 parametrelerini (lambda, augmentation, epoch vb.) gerektiği gibi değiştirebilirsiniz.

```python
import os

os.environ["DEVICE"] = "cuda"  # GPU yoksa CPU moduna düşer
os.environ["RAY_NUM_GPUS"] = "0.2"

# --- PFL PHASE PARAMS ---
os.environ["EPOCHS"] = "10"      # Colab override it as required
os.environ["LAMBDA_P"] = "0.1"   # Target lambda_p
os.environ["AUGMENTATION"] = "1" # Augmentation on

# --- DUMMY DEFENSE ENVIRONMENT CONFIG --- #
os.environ["RAY_TRACING_ENABLED"] = "0"
os.environ["SECURITY_DEFENSES"] = "dummy"

print("Security Defenses:", os.environ.get("SECURITY_DEFENSES"))
print("Epochs:", os.environ.get("EPOCHS"))
print("Lambda_P:", os.environ.get("LAMBDA_P"))
print("Augmentation:", os.environ.get("AUGMENTATION"))
```

## 4) RUN PFL BASELINE (DUMMY DEFENSE)
Çıktılar `outputs/metrics/simulation_results.csv` içerisine yazılır.

```python
import os
import subprocess
import sys

root = os.environ.get("PROJECT_ROOT", "/content/privproject")
print("=== RUNNING DUMMY PROTOTYPE DEFENSE PFL ENGINE ===")
process = subprocess.Popen(
    [sys.executable, "run_pfl.py"],
    cwd=root,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
    bufsize=1,
)

for line in process.stdout:
    print(line, end="", flush=True)

process.wait()
print(f"\nReturn code: {process.returncode}")
```

## 5) Sonuçları indir
`outputs/` klasörünü paketler (güvenlik logları dâhil) ve .zip olarak Colab'dan bilgisayarınıza kaydeder.

```python
import os
import subprocess
from google.colab import files

root = os.environ.get("PROJECT_ROOT", "/content/privproject")
zip_path = "/content/dummy_defense_outputs.zip"
if os.path.isfile(zip_path):
    os.remove(zip_path)

subprocess.check_call(
    ["zip", "-r", zip_path, "outputs", "-x", "*.pyc", "-x", "*__pycache__*"],
    cwd=root,
)
files.download(zip_path)
```
